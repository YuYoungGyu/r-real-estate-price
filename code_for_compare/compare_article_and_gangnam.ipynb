```python
import matplotlib.pyplot as plt
import pandas as pd

path = '../data_for_compare/'

flatten = pd.read_csv(f'{path}flatten_article.csv')
gangnam_trade = pd.read_csv(f'{path}gangnam_trade.csv')
gangnam_rent = pd.read_csv(f'{path}gangnam_rent.csv')
```

```python
x1 = pd.Series([i for i in range(180)])
y1 = flatten['data']
plt.plot(x1, y1)

x2 = [i for i in range(180)]
y2 = [0 for i in range(180)]
plt.plot(x2, y2, color = 'r')

plt.show()
```

```python
plt.plot(gangnam_trade['count'])
```

```python
plt.plot(gangnam_rent['count'])
```

```python
df = pd.DataFrame({"v1" : flatten['data'], "v2" : gangnam_trade['count']})
corr = df.corr(method = 'pearson') 
print("*   ")
display(corr)
```

```python
import seaborn as sns

df_heatmap = sns.heatmap(corr, cbar = True, annot = True, annot_kws={'size' : 20}, fmt = '.2f', square = True, cmap = 'Blues')
```

```python
df = pd.DataFrame({"v1" : list(flatten['data'][60:]), "v2" : list(gangnam_rent['count'])})
corr = df.corr(method = 'pearson') 
print("*   ")
display(corr)

```

```python
import seaborn as sns

df_heatmap = sns.heatmap(corr, cbar = True, annot = True, annot_kws={'size' : 20}, fmt = '.2f', square = True, cmap = 'Blues')
```
